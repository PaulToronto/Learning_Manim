## Running Manim with the Magic Trigger

`%%manim -qm -v WARNING CreateCircle`

1. `%%manim`: The magic trigger.
2. `-qm`: Quality Medium.
    - `-ql` (Low): 480p (Fastest)
    - `-qm` (Medium): 720p (Good balance)
    - `-qh` (High): 1080p (Slowest)
3. `-v WARNING`: Verbosity. This hides the giant wall of text logs (like "MoviePy: Building video...") so your notebook stays clean.
4. CreateCircle: The name of the class you want to render

```python
%%manim -ql -v WARNING CreateCircle

class CreateCircle(Scene):
    def construct(self):
        circle = Circle()
        circle.set_fill(RED, opacity=0.5)
        self.add(circle)
```
